```python
!mysql -u root -e "SHOW DATABASES;"
```

5) CREATE DATABASE

```python
!mysql -u root -D latihan_db -e "CREATE TABLE produk (id INT PRIMARY KEY, nama_produk VARCHAR(50), kategori VARCHAR(30), harga INT);"

#Menampilkan daftar tabel
!mysql -u root -D latihan_db -e "SHOW TABLES;"

#Menampilkan struktur tabel siswa
!mysql -u root -D latihan_db -e "DESCRIBE produk;"
```

6) INSERT INTO

```python
!mysql -u root -D latihan_db -e "INSERT INTO produk (id, nama_produk, kategori, harga) VALUES (1, 'Laptop Asus', 'Elektronik', 8500000), (2, 'Mouse Logitech', 'Aksesoris', 150000), (3, 'Meja Belajar', 'Furniture', 1200000);"
```

7) SELECT

```python
!mysql -u root -D latihan_db -e "SELECT * FROM produk;"
```

8) UPDATE

```python
!mysql -u root -D latihan_db -e "UPDATE produk SET harga = 9000000 WHERE id = 1;"

!mysql -u root -D latihan_db -e "SELECT * FROM produk WHERE id = 1;"
```

9) DELETE

```python
!mysql -u root -D latihan_db -e "DELETE FROM produk WHERE id = 2;"

!mysql -u root -D latihan_db -e "SELECT * FROM produk;"
```
